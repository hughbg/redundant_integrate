## Case 2 Test systematic

Without first.

```python
from integrator import Case2
import matplotlib.pyplot as plt
import numpy as np

num_bl = 50
num_trials = 5000
```

```python
case = Case2(signal=10+10j)
case.print_equation()
case.systematic_cfg = { "axis": "freq", "correlation_fraction": 0, "scale": 1 }
case.track_imaginary(num_trials, num_bl)
```

```python
case = Case2(signal=10+10j)
case.print_equation()
case.systematic_cfg = { "axis": "freq", "correlation_fraction": 0.5, "scale": 1 }
case.track_imaginary(num_trials, num_bl)
```

```python
case = Case2(signal=10+10j)
case.print_equation()
case.systematic_cfg = { "axis": "freq", "correlation_fraction": 1, "scale": 1 }
case.track_imaginary(num_trials, num_bl)
```

```python
lengths = [0, 0.25, 0.5, 1]
print(lengths)
results = []
for correlation_length in lengths:
    case = Case2(signal=10+10j)
    case.systematic_cfg = { "axis": "freq", "correlation_fraction": correlation_length, "scale": 1 }
    x=case.track_imaginary(num_trials, num_bl, stats_only=True)
    results.append(x)
    
print(x.keys())
```

```python
def plot(what):
    values_real = [ r[what+"_real"] for r in results[1:] ]
    values_imag = [ r[what+"_imag"] for r in results[1:] ]
    print(values_real)
    print(values_imag)
    plt.figure(figsize=(12, 6))
    plt.plot(lengths, values_real, label="Real")
    plt.plot(lengths, values_imag, label="Imag")
    if np.min(values_real) > 0 and np.min(values_imag) > 0: plt.ylim(ymin=0)
    plt.title((what if what != "std" else "sigma")+" by correlation length")
    plt.xlabel("Correlation fraction")
    plt.ylabel(what if what != "std" else "sigma")
    plt.legend()
    
def plot_corr():
    values = [ r["correlation"] for r in results[1:] ]
    print(values)
    plt.figure(figsize=(12, 6))
    plt.plot(lengths, values)
    plt.title("Correlation"+" by correlation length")
    plt.xlabel("Correlation fraction")
    plt.ylabel("Correlation")
    plt.legend()

    
plot("mean")
    
```

```python
plot("std")
```

```python
plot("skew")
```

```python
plot("kurtosis")
```

```python
plot_corr()
```
